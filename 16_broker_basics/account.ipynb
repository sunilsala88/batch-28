```python

from alpaca.trading.client import TradingClient
from credentials import api_key,secret_key
trading_client = TradingClient(api_key, secret_key, paper=True)

account = trading_client.get_account()
account
```

```python
import pandas as pd
df=pd.Series(dict(account))
df.to_csv('data.csv')
```

# position

```python
pos=trading_client.get_all_positions()
pos
```

```python
list_of_dict=[]
for i in pos:
    list_of_dict.append(dict(i))
pos_df=pd.DataFrame(list_of_dict)
pos_df.to_csv('position.csv')
pos_df
```

# order book

```python
ord=trading_client.get_orders()
ord
```

```python
list_of_dict=[]
for i in ord:
    list_of_dict.append(dict(i))
ord_df=pd.DataFrame(list_of_dict)
ord_df.to_csv('orders.csv')
ord_df
```
